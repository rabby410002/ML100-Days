## 資料分佈

```python
#Group by
app_train.groupby('CODE_GENDER').count()
```

```python
app_train['CODE_GENDER'].value_counts()
```

```python
## 資料分佈視覺化
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(['F','M'],[app_train['CODE_GENDER'].value_counts()['F'],app_train['CODE_GENDER'].value_counts()['M']])

plt.title("Gender distribution",color="w")
plt.xlabel("Female/Male",color="w")
plt.ylabel("Count",color="w")
#ax.tick_params(axis='x', colors='red')

set_color1=[i.set_color("w") for i in plt.gca().get_xticklabels()]
set_color2=[i.set_color("w") for i in plt.gca().get_yticklabels()]
```
